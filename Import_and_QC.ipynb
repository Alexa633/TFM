```python
# Libreries
#==========
import os, shutil, glob
from pathlib import Path

import pandas as pd
import numpy as np

import anndata
import scanpy as sc
```

```python
#Loop for creating a folder by each bam file using its ID
#========================================================

sourcefile = 'C:/Users/moren/PycharmProjects/TFM/TFM_bulkseq'

# for loop then I split the names of the file then making new folder 
for file_path in glob.glob(os.path.join(sourcefile, '*.bam*')):
    new_dir = file_path.rsplit('.', 1)[0]    
    # If folder does not exist try making new one
    try:
        os.mkdir(os.path.join(sourcefile, new_dir))
    # except error then pass
    except WindowsError:
        pass
    # Move the images from file to new folder based on image nameff
    shutil.move(file_path, os.path.join(new_dir, os.path.basename(file_path)))
```

```python
# Extract each basename for bam files
#=====================================

filenames = []
directory = r'C:\Users\moren\PycharmProjects\TFM\TFM_bulkseq'

for filename in os.listdir(directory):
    
    basename = os.path.splitext (os.path.basename(filename))[0]
    filenames.append (basename)

print(filenames)

#create txt file 

with open("bamfile_list.txt", 'w') as f:
    f.write("\n".join(map(str, filenames)))
```
